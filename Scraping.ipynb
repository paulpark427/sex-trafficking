** Real Scrape Using Selenium **

```python
#get
preference = ['MenSeekWomen', 'MenSeekMen', 'WomenSeekMen', 'WomenSeekWomen', 'Transgender']

pages_urls = []
for i in preference:
    for page in range(1, 2, 1):
        url = "http://atlanta.backpage.com/"+i+"/?page="+str(page)
        pages_urls.append(url)

print len(pages_urls), 'total # of pages'
print "------------------------------------------------------"
pages_urls
```

```python
#beautify ^^^
soups = []
for page in pages_urls:
    html = urllib.urlopen(page).read()
    soup = BeautifulSoup(html, 'html.parser', from_encoding='utf-8')
    soups.append(soup)
    
len(soups)
```

```python
#iterate through soups and grab each link present in each url
url_list = []

for soup in soups:
    for i in soup.find_all('a', href=True):
        if 'atlanta.backpage.com/WomenSeekMen' in i['href']:
            url_list.append(i['href'])
        elif 'atlanta.backpage.com/WomenSeekWomen' in i['href']:
            url_list.append(i['href'])
        elif 'atlanta.backpage.com/MenSeekWomen' in i['href']:
            url_list.append(i['href'])
        elif 'atlanta.backpage.com/MenSeekMen' in i['href']:
            url_list.append(i['href'])
        elif 'atlanta.backpage.com/Transgender' in i['href']:
            url_list.append(i['href'])
        else:
            continue
len(url_list)
```

```python
print len(url_list), 'total # of urls'
print "------------------------------------------------------"

#remove duplicates
url_list = list(set(url_list))
print 'removing the duplicate urls returns', len(url_list), 'urls'
print "------------------------------------------------------"

#removing search urls
for i in url_list:
    if i == 'http://atlanta.backpage.com/MenSeekWomen/':
        url_list.remove(i)
    elif i == 'http://atlanta.backpage.com/MenSeekMen/':
        url_list.remove(i)
    elif i == 'http://atlanta.backpage.com/WomenSeekWomen/':
        url_list.remove(i)
    elif i == 'http://atlanta.backpage.com/WomenSeekMen/':
        url_list.remove(i)
    elif i == 'http://atlanta.backpage.com/Transgender/':
        url_list.remove(i)
    else:
        continue

print 'removing search urls', len(url_list), 'urls'
print "------------------------------------------------------"


#removing urls with page numbers
url_list = [x for x in url_list if 'http://atlanta.backpage.com/Transgender/?' not in x]
url_list = [x for x in url_list if 'http://atlanta.backpage.com/WomenSeekMen/?' not in x]
url_list = [x for x in url_list if 'http://atlanta.backpage.com/WomenSeekWomen/?' not in x]
url_list = [x for x in url_list if 'http://atlanta.backpage.com/MenSeekWomen/?' not in x]
url_list = [x for x in url_list if 'http://atlanta.backpage.com/MenSeekMen/?' not in x]
print 'removing urls that links to other pages return', len(url_list), 'urls'
print "------------------------------------------------------"


#removing urls that link to different areas of atl
for i in url_list:
    if 'Otp' in i:
        url_list.remove(i)
print 'removing urls that link to different areas of atl', len(url_list), 'urls'
print "------------------------------------------------------"

len(url_list)
```

```python
url_list
```

```python
body = []
title = []
location = []
date_time = []

driver = webdriver.Chrome(executable_path="./chromedriver/chromedriver")
for url in url_list:
    

    #go to link
    driver.get(url)

    #wait b/c javascript
    sleep(1)

    #get html
    html = driver.page_source

    #beautiful soup

    #date_time
    soup = BeautifulSoup(html, 'lxml')

    #date_time
    for i in soup.find_all('div', {'id':'pageBackground'}):
        try:
            date_time.append(i.find('div', {'class':'adInfo'}).text)         
        except:
            date_time.append(np.NaN)
            
    #location
    for i in soup.find_all('div', {'class':'posting'}):
        try:
            location.append(i.find('div', {'style':'padding-left:2em;'}).text)
        except:
            location.append(np.NaN)

    #title
    for i in soup.find_all('div', {'id':'pageBackground'}):
        try:
            title.append(i.find('a', {'class':'h1link'}).text)
        except:
            title.append(np.NaN)

    #body
    for i in soup.find_all('div', {'class':'posting'}):
        try:
            body.append(i.find('div', attrs={'class':'postingBody'}).text)
        except:
            body.append(np.NaN)
```

** Clean DataFrame **

```python
for column in df:
    df[column].replace(to_replace=r'\n',value=r'', regex=True, inplace=True)
    df[column].replace(to_replace=r'Location:',value=r'', regex=True, inplace=True)
    df[column].replace(to_replace=r'Posted:',value=r'', regex=True, inplace=True)
    df[column].replace(to_replace=r'\n',value=r'', regex=True, inplace=True)
```
